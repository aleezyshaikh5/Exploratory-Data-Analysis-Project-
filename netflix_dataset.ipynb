# Project Goal:
The goal of this project is to perform Exploratory Data Analysis (EDA) on the Netflix dataset to understand patterns and trends in the platform’s content. The project analyzes information such as movies, TV shows, genres, ratings, release years, duration, directors, actors, and countries.

Main Objectives:
1. Clean and preprocess the Netflix dataset by removing duplicates and handling missing values.
2. Explore the distribution of Movies vs TV Shows on Netflix.
3. Analyze content trends over the years using release dates.
4. Identify top directors, actors, and countries producing the most content.
5. Study the distribution of movie durations.
6. Compare ratings across different categories.
7. Visualize the data using plots such as histograms, bar plots, line plots, and heatmaps.
8. Discover patterns in genres and content production across different countries.

Final Aim:
To gain meaningful insights from the Netflix dataset and understand how content is distributed and produced on the platform through data analysis and visualization.

# Data Visualization

```python
data["Minutes"] = pd.to_numeric(data["Minutes"], errors="coerce")
movie_data = data[(data["Category"] == "Movie") & (data["Minutes"] > 0)]
plt.figure(figsize=(10, 6))
sns.histplot(movie_data["Minutes"].dropna(), bins=30, kde=True, color="skyblue")
plt.title("Distribution of Movie Durations")
plt.xlabel("Minutes")
plt.show()
```

```python
sns.countplot(x="Rating", hue="Category", data=data)
plt.xticks(rotation=45)
plt.title("Rating distribution by Category")
plt.show()
```

```python
top_countries = data["Country"].value_counts().head(10).index
sns.countplot(y="Country", hue="Category", data=data[data["Country"].isin(top_countries)])
plt.title("Top 10 Countries by Number of Movies/TV Shows")
plt.show()
```

```python
year_count = data["Year"].value_counts().sort_index()
year_count.plot(kind="line", marker="o")
plt.title("Number of Releases per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
data["Director"].value_counts().head(15).plot(kind="barh", color="skyblue")
plt.title("Top 15 Directors by Number of Releases")
plt.show()
```

```python
all_casts = data["Cast"].dropna().str.split(", ").explode()
all_casts.value_counts().head(15).plot(kind="barh", color="lightgreen")
plt.title("Top 15 Actors/Actresses")
plt.show()
```

```python
top_types = data["Type"].value_counts().head(20).index
sns.countplot(y="Type", hue="Category", data=data[data["Type"].isin(top_types)])
plt.title("Top 20 Types of Movies vs TV Shows")
plt.show()
```

```python
data["Minutes"] = pd.to_numeric(data["Minutes"], errors="coerce")
plt.figure(figsize=(6,5))
sns.barplot(x="Category", y="Minutes", data=data, ci=None)
plt.title("Average Duration: Movies vs TV Shows")
plt.ylabel("Average Minutes")
plt.show()
```

```python
data.groupby('Year')['Minutes'].max().sort_values(ascending=False).head(10)
```

# Country-Based Genre Analysis

```python
top_types = data['Type'].value_counts().nlargest(15).index
filtered_data = data[data['Type'].isin(top_types)]


final_plot_data = filtered_data[filtered_data["Country"].isin(["United States", "India", "United Kingdom"])]

plt.figure(figsize=(12, 8))
sns.countplot(
    y="Type", 
    hue="Country", 
    data=final_plot_data,
    order=top_types
)

plt.title("Top 15 Genres by Country")
plt.tight_layout()
plt.show()
```
